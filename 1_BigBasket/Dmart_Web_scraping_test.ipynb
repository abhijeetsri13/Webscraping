```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
url = "https://www.dmart.in/category/dry-fruits-aesc-dryfruits"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
data = soup.find("script",attrs={"id": "__NEXT_DATA__"})
if data:
    json_data = json.loads(data.contents[0])
    #print(json_data)
product_data = []

if 'plpData' in json_data['props']['pageProps']:
    products = json_data['props']['pageProps']['plpData']['products']
    for product in products:
        product_name = product['name']
        quantity = product['sKUs'][0]['variantTextValue']
        priceMRP = product['sKUs'][0]['priceMRP']
        priceSALE = product['sKUs'][0]['priceSALE']
        savePrice = product['sKUs'][0]['savePrice']
        variantInfoTxtValue = product['sKUs'][0]['variantInfoTxtValue']

        product_info = {
            'Product Name': product_name,
            'Quantity': quantity,
            'priceMRP': priceMRP,
            'priceSALE': priceSALE,
            'savePrice': savePrice,
            'pricepergm': variantInfoTxtValue
            
        }
        product_data.append(product_info)
        print(product_data)
```

```python
import csv
csv_file = 'dmartproduct_data.csv'

with open(csv_file, 'w', newline='',encoding="utf-8") as csvfile:
    fieldnames = ['Product Name','Quantity', 'priceMRP', 'priceSALE', 'savePrice', 'pricepergm']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for product_info in product_data:
        writer.writerow(product_info)
```
