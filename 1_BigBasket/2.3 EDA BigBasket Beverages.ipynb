### Below function is to seperate the quantiy and its measured unit. using below function we seperated the quantity and it's units.  

```python
import re

def extract_first_numeric_part(quantity_str):

    match = re.match(r'(\d+)\*(\d+)\s*([a-zA-Z]+)', quantity_str)
    match_f = re.match(r'(\d+)\*(\d+\.\d+)\s*([a-zA-Z]+)', quantity_str)
    
    if match:
        multiplier = int(match.group(1))
        factor = int(match.group(2))
        unit = match.group(3)
        print("Multiplier:",multiplier,"factor:",factor,"unit:",unit)

        basic_unit_value = multiplier * factor
        #return basic_unit_value, unit
        new_quant = str(basic_unit_value)+' '+str(unit)
        print('Converted quant:',new_quant)
        return new_quant
    
    elif match_f:
        multiplier = float(match_f.group(1))
        factor = float(match_f.group(2))
        unit = match_f.group(3)
        print("Multiplier_f:",multiplier,"factor_f:",factor,"unit_f:",unit)

        basic_unit_value = multiplier * factor
        #return basic_unit_value, unit
        new_quant = str(basic_unit_value)+' '+str(unit)
        print('Converted quant_f:',new_quant)
        return new_quant
        
    
    else:
        # If the format doesn't match, return the original value
        new_quant = quantity_str
        print("not converted:", new_quant)
        return new_quant
        

BB_Beverages_clean['Quantity_upd'] = BB_Beverages_clean['strip_qua'].apply(extract_first_numeric_part)

# Print the DataFrame with the extracted first numeric parts
print(BB_Beverages_clean[['strip_qua', 'Quantity_upd']])
```

### After cleansing the Quantity column i.e. by removing the unwanted and extra data we converted that into the proper quantity format. We used Regrex function for the same. 
### Converted quantity is pure numeric/float form of the qunatity which is trimmed to remove  

```python
import re

def convert_to_grams_ml(quantity_str):
    # Check if '*' is present indicating multiplication
    # Extract numerical value and unit from the input string
    match = re.match(r'(\d+(\.\d+)?)\s*([a-zA-Z]+)', quantity_str)
        
    if match:
        value = float(match.group(1))
        unit = match.group(3).lower()
            # Convert to grams or milliliters
        if unit in ['g', 'gm', 'grams']:
            return value
        elif unit in ['kg', 'kilograms']:
            return value * 1000
        elif unit in ['l', 'ltr', 'liters']:
            return value * 1000  # Assuming 1 liter is approximately 1000 grams (density varies)
        elif unit in ['ml', 'milliliters']:
            return value
        elif unit in ['pc', 'pcs', 'pieces']:
            return value
        else:
            print(f"Warning: Unknown unit '{unit}'. Assuming it is grams.",quantity_str)
            return value
          
BB_Beverages_clean['Converted_Quantity'] = BB_Beverages_clean['Quantity_upd'].apply(convert_to_grams_ml)

# Print the DataFrame with the converted quantities
print(BB_Beverages_clean[['Quantity', 'Quantity_upd', 'Converted_Quantity']])
```

```python
import numpy as np
BB_Beverages_clean["Price_per_unit(gm/ml/pc)"] = BB_Beverages_clean['Discount_Price'] / BB_Beverages_clean['Converted_Quantity']
BB_Beverages_clean["Price_per_unit(gm/ml/pc)"].round(2)
```
