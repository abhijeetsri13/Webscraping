```python
def convert_to_grams_ml(quantity_str):
    # Check if 'x' is present indicating multiplication
    if 'x' in quantity_str:
        # Use a more flexible regex to handle spaces around 'x'
        match = re.match(r'(\d+(\.\d+)?)\s*[xX]\s*(\d+(\.\d+)?)\s*([a-zA-Z]+)', quantity_str)

        if match:
            value1 = float(match.group(1))
            value2 = float(match.group(3))
            unit = match.group(5).lower()

            result_value = value1 * value2
            return result_value, unit
        else:
            raise ValueError(f"Error: Unable to parse quantity string '{quantity_str}'")

    else:
        # Extract numerical value and unit from the input string
        match = re.match(r'(\d+(\.\d+)?)\s*([a-zA-Z]+)', quantity_str)
        
        if match:
            value = float(match.group(1))
            unit = match.group(3).lower()

            # Convert to grams or milliliters
            if unit in ['g', 'gm', 'grams']:
                return value
            elif unit in ['kg', 'kilograms']:
                return value * 1000
            elif unit in ['l', 'ltr', 'liters']:
                return value * 1000  # Assuming 1 liter is approximately 1000 grams (density varies)
            elif unit in ['ml', 'milliliters']:
                return value
            else:
                print(f"Warning: Unknown unit '{unit}'. Assuming it is grams.",quantity_str)
                return value
        else:
            print(f"Error: Unable to parse quantity string '{quantity_str}'")
            return None

BB_foodnoil_clean['Converted_Quantity'] = BB_foodnoil_clean['strip_qua'].apply(convert_to_grams_ml)

# Print the DataFrame with the converted quantities
print(BB_foodnoil_clean[['Quantity', 'strip_qua', 'Converted_Quantity']])
```
