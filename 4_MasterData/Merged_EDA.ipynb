## Converting price per gram to price per Kg For BigBasket

```python
import pandas as pd
def search_tags(search_string, tags_list):
    return any(tag in search_string for tag in tags_list)
# List of file names
file_names = [r'C:\Users\abhisriv\Documents\webscap\Clean_Dataset\BB_Beverages_clean_7dec.csv', r'C:\Users\abhisriv\Documents\webscap\Clean_Dataset\BB_foodnoil_clean_6Dec.csv', r'C:\Users\abhisriv\Documents\webscap\Clean_Dataset\BB_Snacknbranded_clean_5dec.csv']
tags_list  = ['pcs','teabags']
for file_name in file_names:
    # Load the CSV file into a DataFrame
    df = pd.read_csv(file_name)

    # Function to convert Price_per_unit to Price Per Unit(Kg/L)
    def convert_to_kg_l(row):
        
        result = search_tags(row['Quantity_upd'].lower(), tags_list)
        if result:
            return row['Price_per_unit(gm/ml/pc)']
        else:
            return row['Price_per_unit(gm/ml/pc)'] * (1000)  # Convert grams to kilograms
        
    # Apply the conversion function to create the new column
    df['Price Per Unit(Kg/L)'] = df.apply(convert_to_kg_l, axis=1)

    # Save the updated DataFrame to a new CSV file
    output_file_name = file_name.replace('.csv', '_output.csv')
    df.to_csv(output_file_name, index=False)

    print(f"Conversion completed for {file_name}. Output saved to {output_file_name}")

```

## Converting the merged_output_master.csv to DB

```python
import sqlite3
import pandas as pd
# Function to create a SQLite database and import CSV data
def create_and_load_database(csv_file, db_name='grocery_data.db'):
    # Connect to SQLite database
    conn = sqlite3.connect(db_name)

    # Read CSV into a Pandas DataFrame
    df = pd.read_csv(csv_file)

    # Write the DataFrame to SQLite
    df.to_sql('grocery_data', conn, if_exists='replace', index=False)

    # Commit changes and close the connection
    conn.commit()
    conn.close()

    print(f"Database '{db_name}' created and data loaded successfully.")

# Function to search for a product in the database
def search_product(database, product_name):
    # Connect to SQLite database
    conn = sqlite3.connect(database)

    # Query to search for the product (case-insensitive and partial match)
    query = f"SELECT * FROM grocery_data WHERE \"Product name\" LIKE '%{product_name}%' order by \"Price Per Unit(Kg/L)\""
#SELECT * FROM grocery_data WHERE "Product name" LIKE '%Butter%'
    # Execute the query and fetch the results
    print(query)
    results = pd.read_sql_query(query, conn)

    # Close the connection
    conn.close()

    return results


# Example usage
csv_file_path = r'C:\Users\abhisriv\Documents\webscap\MasterData/merged_output_master.csv'
database_name = 'grocery_data.db'

# Create and load the database
create_and_load_database(csv_file_path, database_name)

# Search for a product
product_name_to_search = 'Biscuit'  # Replace with the product you want to search
search_results = search_product(database_name, product_name_to_search)

# Print the search results
print("Search Results:")
print(search_results)
```

## Creating an app to search Product from the collected (scrapped data)

```python
import tkinter as tk
from tkinter import ttk
import sqlite3
import pandas as pd

class DataSearchApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Data Search App")

        # Create and set up the GUI components
        self.create_widgets()

    def create_widgets(self):
        # Frame for search input
        search_frame = ttk.Frame(self.root, padding="10")
        search_frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

        ttk.Label(search_frame, text="Search:").grid(row=0, column=0, padx=5, pady=5)
        self.search_entry = ttk.Entry(search_frame, width=30)
        self.search_entry.grid(row=0, column=1, padx=5, pady=5)
        self.search_entry.bind("<Return>", lambda event: self.search_data())  # Bind Enter key to search function

        search_button = ttk.Button(search_frame, text="Search", command=self.search_data)
        search_button.grid(row=0, column=2, padx=5, pady=5)

        # Frame for displaying search results
        result_frame = ttk.Frame(self.root, padding="10")
        result_frame.grid(row=1, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

        columns = ['Product name', 'MRP', 'Discount_Price', 'quantity', 'Brand', 'Price Per Unit(Kg/L)',
                   'Total Disc', 'Percentage Disc', 'Platform']

        self.result_tree = ttk.Treeview(result_frame, columns=columns, show='headings')
        for col in columns:
            self.result_tree.heading(col, text=col)
        self.result_tree.grid(row=0, column=0)

        # Frame for displaying suggested items
        suggested_frame = ttk.Frame(self.root, padding="10")
        suggested_frame.grid(row=2, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

        self.suggested_tree = ttk.Treeview(suggested_frame, columns=columns, show='headings')
        for col in columns:
            self.suggested_tree.heading(col, text=col)
        self.suggested_tree.grid(row=0, column=0)

    def search_data(self):
        # Clear previous search and suggested results
        for item in self.result_tree.get_children():
            self.result_tree.delete(item)

        for item in self.suggested_tree.get_children():
            self.suggested_tree.delete(item)

        # Connect to the SQLite database
        conn = sqlite3.connect(r'C:\Users\abhisriv\Documents\webscap\JoiMart\grocery_data.db')

        # Read data based on the search input
        search_input = self.search_entry.get()

        # Split the search input into individual words
        search_terms = search_input.split()

        # Initialize an empty DataFrame to store combined results
        combined_results_df = pd.DataFrame(columns=['Product name', 'MRP', 'Discount_Price', 'quantity', 'Brand', 'Price Per Unit(Kg/L)', 'Total Disc', 'Percentage Disc', 'Platform'])

        # Query each word separately and combine the results
        for term in search_terms:
            query = f"SELECT * FROM grocery_data WHERE [Product name] LIKE '%{term}%'"
            results_df = pd.read_sql_query(query, conn)
            combined_results_df = pd.concat([combined_results_df, results_df], ignore_index=True)

        # Display combined search results in the Treeview
        for index, row in combined_results_df.iterrows():
            self.result_tree.insert("", "end", values=row.tolist())

        # Select the top 5 items with the lowest "Price Per Unit(Kg/L)"
        suggested_items_df = combined_results_df.head(10)
        suggested_items_df = suggested_items_df.nsmallest(5, 'Price Per Unit(Kg/L)')
        # Display suggested items in the Treeview
        for index, row in suggested_items_df.iterrows():
            self.suggested_tree.insert("", "end", values=row.tolist())

        # Close the database connection
        conn.close()

if __name__ == "__main__":
    root = tk.Tk()
    app = DataSearchApp(root)
    root.mainloop()
```
