```python
import pandas as pd
import re

def extract_quantity_unit(row):
    # Check if the text has a multiplication pattern
    multiplication_match = re.match(r'(\d+)x(\d+(\.\d+)?)\s*([a-zA-Z]+)', row['Quantity'].replace(' ','').lower())
    #print(row['Grocery_Name'])
    if multiplication_match:
        factor1, factor2, _, unit = multiplication_match.groups()
        quantity = float(factor1) * float(factor2)
        return quantity, unit
    else:
        # If not a multiplication pattern, use the original pattern
        pattern_match = re.match(r'(\d+\.?\d*)\s*([a-zA-Z]+)', row['Quantity'])
        if pattern_match:
            quantity, unit = pattern_match.groups()
            return float(quantity), unit
        else:
            return None, None

def separate_quantity_units(row):
    # Split the quantity and unit from a single column
    quantity_unit = row['Quantity'].replace('x',' ').split(' ')
    
    if len(quantity_unit) == 2:
        quantity, unit = quantity_unit
        return float(quantity), unit
    else:
        return None, None

def convert_to_kg(quantity, unit):
    # Convert all quantities to kg
    if unit.lower() == 'kg' or unit.lower() =='bags' or unit.lower() =='u' or unit.lower()=='pcs' or unit.lower()== 'cubes' or unit.lower()=='pellets':
        return float(quantity)
    elif unit.lower() == 'g' or unit.lower() == 'gm' or unit.lower() == 'grams' or unit.lower() == 'gram' or unit.lower() == 'gms':
        return float(quantity) / 1000
    elif unit.lower() == 'ml':
        # Assuming 1 ml of water weighs approximately 1 gram
        return float(quantity) / 1000
    elif unit.lower() == 'l':
        return float(quantity) * 1000
    else:
        return None

def calculate_price_per_unit(row):
    # Calculate price per unit in terms of kg
    print(row['Grocery_Name'],"--------------")
    print(row['Quantity'])
    print("Price Per Unit", float(row['Discount_Price']) / convert_to_kg(row['Converted Quantity'], row['Unit']))
    return float(row['Discount_Price']) / convert_to_kg(row['Converted Quantity'], row['Unit'])

def process_csv(input_csv_path, output_csv_path):
    # Read the original CSV file
    df = pd.read_csv(input_csv_path)
    #print(df['Quantity'])
    # Separate quantity and unit into two columns
    df[['Converted Quantity', 'Unit']] = df.apply(extract_quantity_unit, axis=1, result_type='expand')

    # Calculate price per unit
    df['Price Per Unit(Kg)'] = df.apply(calculate_price_per_unit, axis=1)

    # Save the modified DataFrame to a new CSV file
    df.to_csv(output_csv_path, index=False)

def main():
    # Process each CSV file
    input_files = [r'C:\Users\abhisriv\Documents\webscap\Clean_Dataset\dmart_beverages_clean.csv', r'C:\Users\abhisriv\Documents\webscap\Clean_Dataset\dmart_grocery_clean.csv', r'C:\Users\abhisriv\Documents\webscap\Clean_Dataset\dmart_snacks_clean.csv']
    output_files = ['dmart_beverages_processed.csv', 'dmart_grocery_processed.csv', 'dmart_snacks_clean_processed.csv']

    for input_file, output_file in zip(input_files, output_files):
        process_csv(input_file, output_file)

if __name__ == "__main__":
    main()
```

```python
import re

def extract_quantity_unit(text):
    # Check if the text has a multiplication pattern
    multiplication_match = re.match(r'(\d+)x(\d+)\s*([a-zA-Z]+)', text)
    
    if multiplication_match:
        factor1, factor2, unit = multiplication_match.groups()
        quantity = float(factor1) * float(factor2)
        return quantity, unit
    else:
        # If not a multiplication pattern, use the original pattern
        pattern_match = re.match(r'(\d+\.?\d*)\s*([a-zA-Z]+)', text)
        if pattern_match:
            quantity, unit = pattern_match.groups()
            return float(quantity), unit
        else:
            return None, None

# Example usage:
text1 = "1 kg"
quantity1, unit1 = extract_quantity_unit(text1)
print(f"Quantity: {quantity1}, Unit: {unit1}")

text2 = "14x10.8 gm"
quantity2, unit2 = extract_quantity_unit(text2.replace(' ',''))
print(f"Quantity: {quantity2}, Unit: {unit2}")
```
