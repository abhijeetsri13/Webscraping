```python
from bs4 import BeautifulSoup
import requests
import json
import csv
from selenium import webdriver
import time
import datetime
import os
```

```python
url_list=[]
url_list.append("https://www.dmart.in/category/grocery-aesc-grocerycore")
url_list.append("https://www.dmart.in/category/dmart-grocery-aesc-dmartgrocerycore")
url_list.append("https://www.dmart.in/category/ready-to-cook-aesc-readytocook")
url_list.append("https://www.dmart.in/category/packaged-food-aesc-packagedfoodcore")
url_list.append("https://www.dmart.in/category/beverages-aesc-beverages")
for x in url_list:
    driver = webdriver.Chrome()
    driver.get(x)
    driver.implicitly_wait(5)
    screen_height = driver.execute_script("return window.screen.height;")  # Browser window height
    # Automatically scroll the page
    scroll_pause_time = 3  # Pause between each scroll
    screen_height = driver.execute_script("return window.screen.height;")  # Browser window height
    i = 1
    scroll = 1
    while True:
        # Scroll down
        driver.execute_script(f"window.scrollTo(0, {screen_height * i});")
        i += 1
        time.sleep(scroll_pause_time)
        scroll = scroll + 1
        print('scroll:',scroll)
    
        # Check if reaching the end of the page
        scroll_height = driver.execute_script("return document.body.scrollHeight;")
        if screen_height * i > scroll_height or scroll == 150:
            break
    
    soup = BeautifulSoup(driver.page_source, "html.parser")
    #print(soup)
    filename = x.replace("/","_").replace(":","_")
    txt_file = filename+".txt"
    prettyf = open(txt_file, "wb")
    prettyf.write(soup.prettify().encode("utf-8"))
    prodlist =soup.find_all(class_="MuiGrid-root MuiGrid-item MuiGrid-grid-xs-12 MuiGrid-grid-sm-6 MuiGrid-grid-md-4 MuiGrid-grid-lg-auto MuiGrid-grid-xl-auto mui-style-tlgnrr")
    product_data = []
    for prod in prodlist:
        gr = prod.find_all(class_='vertical-card_title__awihj')
 
        for h in gr:
            Grocery_Name = h.get_text().strip()
            print("Grocery_Name=",Grocery_Name)
            brand_name = Grocery_Name.split(" ")[0].strip()
            print("brand_name=",brand_name)
            mrplist = prod.find_all(class_='vertical-card_amount__muVeb',style="text-decoration: line-through;")
            for mrp in mrplist:
                mrp_Price = mrp.get_text().strip()
                print('MRP_Price=',mrp_Price)
            offlist = prod.find_all(class_='vertical-card_section-right__SEVEZ')
            for off in offlist:
                off_Price = off.find(class_='vertical-card_amount__muVeb').get_text().strip()
                if off_Price != '--':
                    print('off_Price=',off_Price)
                    m = float(mrp_Price)
                    o = float(off_Price)
                    dis_Price = m - o
                    print('dis_Price=',dis_Price)
                else:
                    dis_Price = 0
                    print('dis_Price=',dis_Price)
            quantity = prod.find_all(class_='bootstrap-select_option__tODFG')
            for q in quantity:
                quant = q.get_text().strip()
                quantity = quant.split("(")[0].strip()
                print('quantity=',quantity)
            priceperunit = prod.find_all(class_='bootstrap-select_infoTxt-value__2Ywcm')
            for pu in priceperunit:
                pr_un = pu.get_text().strip()
                price_unit = pr_un.split("/")[0].strip()
                price_unit = price_unit.replace("(","")
                print('price_per_unit=',price_unit)
    
        product_info = {
            'Grocery_Name': Grocery_Name,
            'MRP_Price' : mrp_Price,
            'Quantity' : quantity,
            'Discount_Price': dis_Price,
            'brand_name' : brand_name,
            'price_per_unit' : price_unit
                
        }
        product_data.append(product_info)
        
        csv_file = filename+".csv"
    
        with open(csv_file, 'w', newline='',encoding="utf-8") as csvfile:
             fieldnames = ['Grocery_Name','MRP_Price','Discount_Price','Quantity','brand_name','price_per_unit']
             writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
             writer.writeheader()
             for product_info in product_data:
                writer.writerow(product_info)
    prettyf.close()
    
        
    driver.close()
```
